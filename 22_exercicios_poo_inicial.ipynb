# Exercícios de mosaico de gráficos

Considere os dados da primeira célula para um aplicativo de uma pequena empresa de software. Faça um mosaico com 4 gráficos:

1. Gráfico de linhas com os novos usuários por mês
2. Gráfico de barras com a receita mensal
3. Gráfico de linhas com os usuários ativos diários
4. Gráfico de barras empilhadas com os tickets de suporte abertos e fechados

Os gráficos 1 e 2 devem ter uma linha própria no mosaico. Os gráficos 3 e 4 devem compartilhar a mesma linha.


```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 140, 150, 160, 170, 120, 110, 150, 200, 140]
usuarios_ativos_diarios = [1000, 750, 900, 1300, 1400, 1500, 1200, 1200, 1500, 1900, 2000, 1800]
receita_mensal = [10000, 8500, 9800, 10000, 11000, 12000, 10000, 10500, 13000, 13000, 20000, 17000]
tickets_suporte_abertos = [110, 75, 100, 140, 170, 160, 100, 130, 190, 200, 210, 140]
tickets_suporte_fechados = [100, 85, 100, 120, 130, 150, 140, 110, 130, 120, 150, 200]
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
plt.rc('axes', prop_cycle=(cycler('color', plt.get_cmap('Pastel1').colors)))

fig, axs = plt.subplot_mosaic('aa;bb;cd', figsize=(14,12), gridspec_kw={'wspace':0.2, 'hspace':0.4})

#fig1:

axs['a'].plot(meses, novos_usuarios, linewidth=3, marker='o', markersize=8)
axs['a'].tick_params(axis='x', length=0)

for i, marcador in enumerate(meses):
    axs['a'].annotate(novos_usuarios[i], xy=(meses[i], novos_usuarios[i]), xytext=(-7,6), textcoords='offset points', weight='bold')

axs['a'].set_title('Número de Novos Usuários por Mês')
axs['a'].set_ylim((50,250))
axs['a'].set_yticks([])
axs['a'].set_frame_on(False)

#fig2:

barra = axs['b'].bar(meses, receita_mensal, edgecolor='k')
axs['b'].bar_label(barra, labels=receita_mensal, weight='bold', padding=3)

axs['b'].set_ylim((0,25000))
axs['b'].set_yticks([])
axs['b'].tick_params(axis='x', length=0)
axs['b'].set_title('Receita Mensal (em mil)')
axs['b'].set_frame_on(False)

#fig3:

axs['c'].plot(meses, usuarios_ativos_diarios, linewidth=3, marker='o', markersize=8)
axs['c'].tick_params(axis='x', length=0)

for i, marcador in enumerate(meses):
    axs['c'].annotate(usuarios_ativos_diarios[i], 
xy=(meses[i], usuarios_ativos_diarios[i]), xytext=(-7,6), textcoords='offset points', weight='bold', fontsize=7.5)

axs['c'].set_title('Número de Usuários Ativos por Mês')
axs['c'].set_ylim((500, 2500))
axs['c'].set_yticks([])
axs['c'].set_frame_on(False)

#fig4:

aberto = axs['d'].bar(meses, tickets_suporte_abertos, label='Aberto', edgecolor='k')
fechado = axs['d'].bar(meses, tickets_suporte_fechados, bottom=tickets_suporte_abertos, label='Fechado', edgecolor='k')
axs['d'].bar_label(aberto, labels=tickets_suporte_abertos, label_type='center', weight='bold', fontsize=8)
axs['d'].bar_label(fechado,labels=tickets_suporte_fechados, label_type='center', weight='bold', fontsize=8)

axs['d'].tick_params(axis='x', length=0)
axs['d'].legend()
axs['d'].set_title('Tickets de Suporte Mensais')
axs['d'].set_ylim((0, 450))
axs['d'].set_frame_on(False)

plt.suptitle('Relatório de Desenpenho Anual', weight='bold', fontsize=16)
plt.show()
```
